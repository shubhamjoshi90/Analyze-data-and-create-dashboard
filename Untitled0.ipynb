```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
test_df = pd.read_csv(r'/content/test.csv')
print(test_df.shape)
test_df.head()
```

```python
test_df.info()
```

```python
test_df = test_df.astype('float32')
```

```python
train_df = pd.read_csv(r'/content/train.csv')
print(train_df.shape)
train_df.head()
```

```python
train_df.info()
```

```python
train_df = train_df.astype('float32')
```

```python
# Correlation with target, FloodProbability
corrmat = train_df.drop(['id'], axis=1).corr()
# corrmat[['FloodProbability']]
corrmat['FloodProbability']
```

```python
# correlation plot
plt.figure(figsize=(12, 12))
k=21
cols = corrmat.nlargest(k, 'FloodProbability')['FloodProbability'].index
cm = np.corrcoef(train_df[cols].values.T)
sns.set(font_scale=0.8)
hm = sns.heatmap(cm, cbar=False, annot=True, square=True, fmt='.3f', annot_kws={'size': 9},
                 yticklabels=cols.values, xticklabels=cols.values, cmap="Purples")
hm.xaxis.tick_top()
plt.xticks(rotation=45, ha='left')
plt.show()
```

```python
# only for feature
feature = train_df.drop(['id', 'FloodProbability'], axis=1)
target = train_df[['FloodProbability']]
print(feature.shape)
feature.head()
```

```python
target
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
feature_std = StandardScaler().fit_transform(feature)

feature_std = pd.DataFrame(feature_std, columns=feature.columns)
feature_std.head(2)
```

```python
# 0.1% sampling
feature_std_smp = feature_std.sample(frac=0.001)
feature_std_smp.shape
```

```python
feature_std_smp_T = feature_std_smp.T
feature_std_smp_T.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
%%time
clusters = linkage(y=feature_std_smp_T, method='ward', metric='euclidean')
# single, complete, average, centroid, ward
clusters.shape
```

```python
plt.figure( figsize = (16, 4) )
dendrogram(clusters, leaf_font_size=12, labels= feature_std_smp.columns,
           color_threshold=49)
plt.axhline(y = 49, color = 'r', linestyle = '-')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster # 지정한 클러스터 자르기

cut_tree = fcluster(clusters, t=49, criterion='distance')
cut_tree # prediction
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(feature_std, target, random_state = 2405)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
# xgboost
import xgboost as xgb
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

xgbr = xgb.XGBRegressor()

model_xgbr = xgbr.fit(X_train, y_train)
model_xgbr_pred = model_xgbr.predict(X_val)
print('model_xgbr R2_Score: {0:.4f}'.format(r2_score(y_val, model_xgbr_pred)))
print('model_xgbr MAE: {0:.4f}'.format(mean_absolute_error(y_val, model_xgbr_pred)))
print('model_xgbr MSE: {0:.4f}'.format(mean_squared_error(y_val, model_xgbr_pred)))
```

```python
# LightGBMRegressor
from lightgbm import LGBMRegressor

lgbmr = LGBMRegressor(verbose=-1)

model_lgbmr = lgbmr.fit(X_train, y_train)
model_lgbmr_pred = model_lgbmr.predict(X_val)
print('model_lgbmr R2_Score: {0:.4f}'.format(r2_score(y_val, model_lgbmr_pred)))
print('model_lgbmr MAE: {0:.4f}'.format(mean_absolute_error(y_val, model_lgbmr_pred)))
print('model_lgbmr MSE: {0:.4f}'.format(mean_squared_error(y_val, model_lgbmr_pred)))
```

```python
# CatBoostRegressor
from catboost import CatBoostRegressor

cbr = CatBoostRegressor(verbose=0)

model_cbr = cbr.fit(X_train, y_train)
model_cbr_pred = model_cbr.predict(X_val)
print('model_cbr R2_Score: {0:.4f}'.format(r2_score(y_val, model_cbr_pred)))
print('model_cbr MAE: {0:.4f}'.format(mean_absolute_error(y_val, model_cbr_pred)))
print('model_cbr MSE: {0:.4f}'.format(mean_squared_error(y_val, model_cbr_pred)))
```

```python
# weight model blending
wmb_pred = (model_xgbr.predict(X_val)*3/10 +
            model_lgbmr.predict(X_val)*1/10 +
            model_cbr.predict(X_val)*6/10)
print('wmb_pred R2_Score: {0:.4f}'.format(r2_score(y_val, wmb_pred)))
print('wmb_pred MAE: {0:.4f}'.format(mean_absolute_error(y_val, wmb_pred)))
print('wmb_pred MSE: {0:.4f}'.format(mean_squared_error(y_val, wmb_pred)))
```

```python
plt.figure(figsize=(16, 3))

plt.subplot(1, 4, 1)
plt.scatter(x = y_val, y = model_xgbr_pred, color ='green')
plt.plot([y_val.min(), y_val.max()], [model_xgbr_pred.min(), model_xgbr_pred.max()])
plt.title('model_xgbr')
plt.xlabel('y_val', fontsize = 9)
plt.ylabel('xgbr_pred', fontsize = 9)

plt.subplot(1, 4, 2)
plt.scatter(x = y_val, y = model_lgbmr_pred, color ='green')
plt.plot([y_val.min(), y_val.max()], [model_lgbmr_pred.min(), model_xgbr_pred.max()])
plt.title('model_lgbmr')
plt.xlabel('y_val', fontsize = 9)
plt.ylabel('lgbmr_pred', fontsize = 9)

plt.subplot(1, 4, 3)
plt.scatter(x = y_val, y = model_cbr_pred, color ='green')
plt.plot([y_val.min(), y_val.max()], [model_cbr_pred.min(), model_cbr_pred.max()])
plt.title('model_cbr')
plt.xlabel('y_val', fontsize = 9)
plt.ylabel('catr_pred', fontsize = 9)

plt.subplot(1, 4, 4)
plt.scatter(x = y_val, y = wmb_pred, color ='green')
plt.plot([y_val.min(), y_val.max()], [wmb_pred.min(), wmb_pred.max()])
plt.title('model blending')
plt.xlabel('y_val', fontsize = 9)
plt.ylabel('wmb_pred', fontsize = 9)

plt.show()
```
